```python
import pyleoclim as pyleo
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import ammonyte as amt
import seaborn as sns
import pandas as pd
```

```python
def transition_detect(series):
    '''Function to detect transitions across a boundary threshold
    
    Parameters
    ----------
    
    series : pyleoclim.Series or ammonyte.Series
        Series to detect transitions
        
    Returns
    -------
    
    transitions : list
        List of transition timings'''

    series_fine = series.interp(step=1)

    upper, lower = amt.utils.sampling.confidence_interval(series)

    above_thresh = np.where(series_fine.value > upper,1,0)
    below_thresh = np.where(series_fine.value < lower,1,0)

    transition_above = np.diff(above_thresh)
    transition_below = np.diff(below_thresh)

    upper_trans = series_fine.time[1:][np.diff(above_thresh) != 0]
    lower_trans = series_fine.time[1:][np.diff(below_thresh) != 0]

    full_trans = np.zeros(len(transition_above))

    last_above = 0
    last_below = 0
    for i in range(len(transition_above)):
        above = transition_above[i]
        below = transition_below[i]
        if above != 0:
            if last_below+above == 0:
                loc = int((i+below_pointer)/2)
                full_trans[loc] = 1
                last_below=0
            last_above = above
            above_pointer = i
        if below != 0:
            if last_above + below == 0:
                loc = int((i+above_pointer)/2)
                full_trans[loc] = 1
                last_above=0
            last_below = below
            below_pointer = i

    transitions = series_fine.time[1:][full_trans != 0]
    
    return transitions
```

```python
color_list = sns.color_palette('colorblind')
```

```python
series_name_list = ['NGRIP.NGRIP.2004',
                   'Renland.Johnsen.1992',
                   'GRIP.Vinther.2006',
                   'EPICADomeC.Stenni.2010']

index_list = [0,2,2,5]
```

```python
series_list = []

for idx,name in enumerate(series_name_list):
    path=f'./8k_ice/{name}.lpd'
    lipd = pyleo.Lipd(path)
    lipd_list = lipd.to_LipdSeriesList()
    series_list.append(lipd_list[index_list[idx]])
    
gisp = pd.read_csv('./8k_ice/GISPd18O.csv',na_values=999999)
gisp_series = pyleo.Series(
    time=gisp['Age [yr BP]'],
    value=gisp['d18O [permil]'],
    time_name = 'Age',
    time_unit = 'kyr BP',
    value_name='d18O',
    value_unit = 'permil',
    label='GISP')

series_list.append(gisp_series)
    
ms_ice = pyleo.MultipleSeries(series_list)
```

```python
ms_ice.stackplot()
```

```python
m = 12
end_list = np.arange(11000,11640,40)
res_dict = {}
lp_dict = {}

for idx,series in enumerate(ms_ice.series_list):
    name = series.label
    print(f'Analyzing {name}')
    for end_time in tqdm(end_list):
        series_slice = series.slice((0,end_time))
        td = amt.TimeEmbeddedSeries(series.slice((0,end_time)),m)
        eps = td.find_epsilon(1,.05,.01,verbose=False)
        rm = eps['Output']
        lp = rm.laplacian_eigenmaps(w_size=20,w_incre=4)
        transitions = transition_detect(lp)
        
        if name not in res_dict:
            res_dict[name] = []
        if name not in lp_dict:
            lp_dict[name] = []
            
        res_dict[name].append(transitions)
        lp_dict[name].append(lp)
```

```python
name = series_name_list[0]

fig,ax = plt.subplots(nrows=len(lp_dict[name]),figsize=(20,40),sharex=True)
fig.subplots_adjust(hspace=1)
axes = ax.ravel()

for idx,series in tqdm(enumerate(lp_dict[name])):
    series.confidence_fill_plot(ax=axes[idx],title=f'Time end = {end_list[idx]}')
    
fig.suptitle(name)
```
